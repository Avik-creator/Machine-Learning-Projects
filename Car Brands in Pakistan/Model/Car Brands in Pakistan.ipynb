## Importing the Required Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('data.csv')
```

## Changing of DataType for Better Visualization of Data.

```python
df['Max Price'] = df['Max Price'].str.replace(',', '').astype(int)
df['Max Price'] = pd.to_numeric(df['Max Price'])
```

```python
df['Min Price'] = df['Min Price'].str.replace(',', '').astype(int)
df['Min Price'] = pd.to_numeric(df['Min Price'])
```

```python
df.describe()
```

```python
df.isna().sum()
```

## Visualisation of Data

```python
plt.figure(figsize=(20, 10))
df.groupby('Model')['Max Price'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Model')['Min Price'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Make')['Max Price'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Make')['Min Price'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Make')['Year'].mean().sort_values(ascending=False).plot(kind='bar')
```

## Plotting the Above data in Pie Chart

```python
plt.figure(figsize=(10, 10))
top = df['Min Price'].value_counts().sort_values(ascending=False).head(6)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(10, 10))
top = df['Make'].value_counts().sort_values(ascending=False).head(6)
top.plot(kind='pie')
```
