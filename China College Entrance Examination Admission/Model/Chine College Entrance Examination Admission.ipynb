```python
df.columns = ['Birth Year', 'College Entrance Examination Year', 'NewBorn Population', 'University Admission Expansion Policy (Boolean)','GDP in Birth Year',
             'GDP in College Entrance Year', 'Amount of Participants in the College Entrance Examination',
             'Amount of admission of the college entrance examination', 'College entrance examination admission rate',
             'Gross enrollment rate']
```

### 3.5 Clearing Dataset of Null values

### 3.6 Seeing if null Values is present. 

```python
df1.isna().sum()
```

## 4. Data Visualisation

### 4.1 Bar Plots

```python
df1['Amount of Participants in the College Entrance Examination'].value_counts().plot.bar(title='Number of Applicants',color='lightblue')
```

```python
df1['Amount of admission of the college entrance examination'].value_counts().plot.bar(title='Admission Rate',color='lightblue')
```

```python
plt.figure(figsize=(20, 10))
df1.groupby('Amount of Participants in the College Entrance Examination')['Amount of admission of the college entrance examination'].mean().sort_values(ascending=True).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df1.groupby('College Entrance Examination Year')['Amount of Participants in the College Entrance Examination'].mean().sort_values(ascending=True).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df1.groupby('College Entrance Examination Year')['Amount of admission of the college entrance examination'].mean().sort_values(ascending=True).plot(kind='bar')
```

### 4.2 Pie Chart

```python
plt.figure(figsize=(10, 10))
plt.pie(x=df1['Amount of admission of the college entrance examination'].sort_values(ascending=True).head(),
        labels=df1['College Entrance Examination Year'].sort_values(ascending=True).head(),
        autopct='%2.1f%%')

plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(x=df1['Amount of admission of the college entrance examination'].sort_values(ascending=True).tail(),
        labels=df1['College Entrance Examination Year'].sort_values(ascending=True).tail(),
        autopct='%2.1f%%')

plt.show()
```

### 4.3 Plotting against Every Column

```python
sns.pairplot(df1)
```
